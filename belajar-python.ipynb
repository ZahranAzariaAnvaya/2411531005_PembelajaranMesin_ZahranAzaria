```python
# IF ELIF dan ELSE
def nilai_akhir(nilai):
    if nilai >= 80:
        predikat = "A"
    elif nilai >= 70:
        predikat = "B"
    elif nilai >= 60:
        predikat = "C"
    else:
        predikat = "D"
    return predikat
    
hasil = nilai_akhir(76)
print("Nilai akhir adalah", hasil)
```

```python
# Membuat dan mencetak hasil sebuah fungsi 
def luas_persegi(sisi):
    luas = sisi * sisi
    return luas

hasil = luas_persegi(4)
print("Luas persegi adalah:", hasil)
```
